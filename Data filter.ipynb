# Combine data and reorder column

```python
unified_col = ['user_id',
               'display_name',
               'gender',
               'location',
               'badge_counts',
               'reputation',
               'accept_rate',
               'question_count',
               'answer_count',
               'up_vote_count',
               'down_vote_count',
               'account_id',
               'user_type']
unified_col
```

1. firstk = 1 - 1000
2. secondk = 1001 - 2000

```python
import pandas as pd
data_firstk = pd.read_csv("raw_user_data/page_1_to_1000.csv")
data_firstk = data_firstk[unified_col]
data_firstk
```

```python
data_secondk = pd.read_csv("raw_user_data/page_1001_to_2000.csv")
# secondk_col = list(data_secondk.columns.values)
# secondk_col
data_secondk = data_secondk[unified_col]
data_secondk
```

```python
data_fourthk = pd.read_csv("raw_user_data/page_4001_to_5000.csv")
data_fourthk = data_fourthk[unified_col]
data_fourthk
```

```python
data_total= pd.concat([data_firstk,data_secondk,data_fourthk])
data_total
```

# get male and female data

```python
female_total = data_total.query("gender == 'female'")
female_total = female_total.drop_duplicates(subset='user_id')
female_total = female_total.reset_index()
female_total = female_total.drop(columns=['index'])
female_total
```

```python
male_total = data_total.query("gender == 'male'")
male_total = male_total.drop_duplicates(subset='user_id')
male_total = male_total.reset_index()
male_total = male_total.drop(columns=['index'])
male_total
```

```python
from sklearn.utils import shuffle

male_sample = shuffle(male_total,n_samples=female_total.shape[0])
male_sample = male_sample.reset_index()
male_sample = male_sample.drop(columns=['index'])
male_sample
```

```python
female_total.to_csv('raw_user_data/user_female_total.csv',index=None)
male_total.to_csv('raw_user_data/user_male_total.csv',index=None)
male_sample.to_csv('raw_user_data/user_male_sample.csv',index=None)
```
